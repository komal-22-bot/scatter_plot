```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('googleplaystore.csv',sep='\t')
print(df.columns)
df_paid=df[df['Paid/Free']=='Paid']
```

```python
sns.scatterplot(data=df_paid, x='Installs', y='Revenue', hue='Category')
plt.title('Scatter Plot of Revenue vs. Installs for Paid Apps')
plt.xlabel('Number of Installs')
plt.ylabel('Revenue')
plt.show()
```

```python
sns.lmplot(x='Installs', y='Revenue', hue='Category', data=df_paid, ci=None)
plt.title('Scatter Plot with Trendline of Revenue vs. Installs for Paid Apps')
plt.xlabel('Number of Installs')
plt.ylabel('Revenue')
plt.show()
```
